```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

%matplotlib inline
```

```python
pd.pandas.set_option('display.max_columns', None)
pd.pandas.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
df.head()
```

```python
df1 = df.copy()
df1 = df1.drop(['Id','SalePrice'],axis=1)

df2 = test.copy()
df2 = df2.drop('Id',axis=1)
```

```python
df_comb = pd.concat([df1,df2],ignore_index=True, axis=0)
```

```python
df_comb.tail()
```

```python
df_comb.isnull().sum()
```

```python
df_comb.drop(['Alley','PoolQC','Fence','MiscFeature','FireplaceQu'], axis=1, inplace=True)
```

```python
numerical_features = [features for features in df_comb.columns if df_comb[features].dtype != 'O']

print(f'total {len(numerical_features)} numerical features')
```

```python
categorical_features = [features for features in df_comb.columns if df_comb[features].dtype == 'O']

print(f'total {len(categorical_features)} categorical features')
```

```python
# EDA of categorical varibles

for feature in categorical_features:
    df.groupby(feature)['SalePrice'].median().plot.bar()
    plt.show()
```

```python
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
len(year_feature)
```

```python
# EDA of Year features
for feature in year_feature:
    plt.scatter(df1[feature],df['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('Sale Price')
    plt.show()
```

```python
pd.plotting.scatter_matrix(df_comb[year_feature],figsize=(10,10));
```

```python
discrete_features = [feature for feature in numerical_features if df_comb[feature].nunique() < 39 and feature not in year_feature]
len(discrete_features)
```

```python
df_comb[numerical_features].nunique()
```

```python
# EDA of discrete features

for feature in discrete_features:
    df.groupby(feature)['SalePrice'].median().plot.bar()
    plt.show()
```

```python
pd.plotting.scatter_matrix(df_comb[discrete_features],figsize=(15,15));
```

```python
continuous_features = [feature for feature in numerical_features if feature not in discrete_features+year_feature]
len(continuous_features)
```

```python
ordinal_features = ['GarageQual', 'GarageCond', 'GarageFinish', 'KitchenQual', 'HeatingQC', 'BsmtExposure', 'BsmtCond', 'BsmtQual', 'ExterCond', 'ExterQual']
```

```python
for i in ordinal_features:
    categorical_features.remove(i)
```

```python
len(categorical_features)
```

```python
df_comb[continuous_features] = df_comb[continuous_features].apply(lambda x: np.log(1+x))
```

```python
for feature in continuous_features:
    plt.hist(df_comb[feature],bins=20)
    plt.title(feature)
    plt.show()
    
```

```python
pd.plotting.scatter_matrix(df_comb[continuous_features],figsize=(15,15));
```

```python
for feature in year_feature:
    if feature == 'YrSold':
        pass
    else:
        df_comb[feature] = df_comb['YrSold'] - df_comb[feature]
```

```python
df_comb[categorical_features].isnull().sum()
```

```python
for feature in year_feature:
    if df_comb[feature].isnull().sum() > 0:
        df_comb[feature+'_na'] = np.where(df_comb[feature].isnull(), 1, 0)
        
```

```python
df_comb['LotFrontage_na'] = np.where(df_comb['LotFrontage'].isnull(), 1, 0)
df_comb['MasVnrArea_na'] = np.where(df_comb['MasVnrArea'].isnull(), 1, 0)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ode = OrdinalEncoder()
```

```python
ohe = OneHotEncoder(handle_unknown='ignore',sparse=False)
```

```python
yr_transformer = Pipeline(steps=[('yr',SimpleImputer(strategy='median')),
                                 ('yr_scaling',MinMaxScaler())])
```

```python
conti_transformer = Pipeline(steps=[('cont',SimpleImputer(strategy='median')),
                                    ('cont_scaling',MinMaxScaler())])
```

```python
disc_transformer = Pipeline(steps=[('disc', SimpleImputer(strategy='most_frequent')),
                                   ('scaler',MinMaxScaler())])
```

```python
ordi_transformer = Pipeline(steps = [('ord', SimpleImputer(strategy='most_frequent')),
                                     ('ord1', ode),
                                     ('scaler', MinMaxScaler())])
```

```python
category_trans = Pipeline(steps =  [('cat',SimpleImputer(strategy='constant',fill_value='missing')),
                                    ('cat1',ohe)])
```

```python
preprocessor = ColumnTransformer(transformers = [('yr_trans',yr_transformer,year_feature),
                                                 ('cont_trans',conti_transformer,continuous_features),
                                                 ('disc_trans',disc_transformer,discrete_features),
                                                 ('ordi_trans', ordi_transformer, ordinal_features),
                                                 ('cate_trans',category_trans,categorical_features)],
                                 remainder='passthrough',
                                 sparse_threshold=0.0)
```

```python
df_comb = preprocessor.fit_transform(df_comb)
```

```python
df_comb.shape
```

```python
cat_column_name = preprocessor.transformers_[4][1]\
                   .named_steps['cat1'].get_feature_names(categorical_features)

```

```python
columns = []

columns.extend(year_feature)

columns.extend(continuous_features)

columns.extend(discrete_features)

columns.extend(ordinal_features)

columns.extend(cat_column_name)

passthrough_columns = ['GarageYrBlt_Na','LotFrontage_Na','MasVnrArea_na']

columns.extend(passthrough_columns)
```
